### Data Pre-Processing

#### Since we have Claimant DOB I am removing the Claimant Date Of Birth column

```python
df.drop('Claimant Date Of Birth', axis=1, inplace = True)
```

#### Changing the Claim status, Death_Claim, and Claimant_Sex for easier understanding.

```python
df['Claim Status'].value_counts()
```

```python
df['Death_Claim'].value_counts()
```

```python
df['Claimant_Sex'].value_counts()
```

- I am going to change these values into more understandable values.
- There are also some values which are not eplained in the Claimant_Sex, so I am changing them into 'Unknown'

```python
df['Claim Status'] = np.where(df['Claim Status'] == 'C', 'Closed', df['Claim Status'])
df['Claim Status'] = np.where(df['Claim Status'] == 'O', 'Open', df['Claim Status'])

df['Death_Claim'] = np.where(df['Death_Claim'] == 'Y', 'Yes', df['Death_Claim'])
df['Death_Claim'] = np.where(df['Death_Claim'] == 'N', 'No', df['Death_Claim'])

df['Claimant_Sex'] = df['Claimant_Sex'].map(
    lambda x: 'Female' if x == 'F' else ('Male' if x == 'M'  else 'Unknown'))
```

```python
df['Claimant_Sex'].value_counts()
```

```python
df.info()
```

#### Changing the Date columns from object into Datetime columns. 

```python
df['Claimant DOB'] = pd.to_datetime(df['Claimant DOB']).dt.date
df['Accident Date'] = pd.to_datetime(df['Accident Date']).dt.date
df['Claimant DOB'] = pd.to_datetime(df['Claimant DOB']).dt.date
df['Claim Closed Date'] = pd.to_datetime(df['Claim Closed Date']).dt.date
```

```python
df['Claimant Age'] =round((df['Accident Date']- df['Claimant DOB'] )/ np.timedelta64(1, 'Y'),0)
```
